# Importing Lib's

```python
import numpy as np
import pandas as pd
from transformers import pipeline
```

```python
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')
```

```python
test_df.head()
```

```python
train_df.shape, test_df.shape
```

# Pipeline

```python
classf = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

# Traning

```python
# Traning whole test_data on pipeline model.
seq = test_df['Description'].tolist()
labels = ["business", "technology", "politics", "sports"]

classification_results = classf(seq, labels)
```

```python
classification_results
```
