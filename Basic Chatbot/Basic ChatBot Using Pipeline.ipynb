```python
from transformers import pipeline
```

# Using GPT2

```python
chatbot = pipeline('text-generation', model= 'microsoft/DialoGPT-small')
chat_history = ''
print("ChatBot is Ready to Use. Type 'quit' to exit!")

while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        break

    chat_history += f"User: {user_input}\nBot:" # Interface to show User: & Bot:

    response = chatbot(
        chat_history,
        max_new_tokens=100,
        pad_token_id=50256, # GPT2/DialoGPT
        do_sample=True, # Allow randomness
        top_k=50, # only took top 50 words
        top_p=0.95, # take words that cover 95%.
    )

    generated_text = response[0]["generated_text"]
    bot_reply = generated_text[len(chat_history):].split("\n")[0] # removes the old chat history but keeps the new ones.
    print(f"Bot: {bot_reply}")

    chat_history += f" {bot_reply}\n"
```
